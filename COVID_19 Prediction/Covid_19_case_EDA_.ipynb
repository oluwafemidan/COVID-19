```python
# Import the required Libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Data
df = pd.read_csv('/content/COVID19 - COVID19.csv')
pd.set_option('display.max_columns', None)
df.head(9)
```

# Exploratory Data Analysis(EDA)

```python
#getting the dimensions of a DataFrame in pandas;(rows,columns).
df.shape
```

```python
#checking the column names
df.columns
```

```python
# Rename columns (replace spaces with underscores)
df.columns = df.columns.str.replace(" ", "_")

df.columns
```

```python
df.info()
```

```python
#checking for missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
#checking for missing values in percentage
(df.isnull().mean().sort_values(ascending=False) * 100).round(2)
```

```python
#since the data contain lot of missin value; columns having more than 70% of missing values is not worth keeping, so dropping the column will be better.
to_drop = df.columns[df.isnull().mean() > 0.7]
print("Columns to drop:", list(to_drop))
df = df.drop(columns=to_drop)
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
#checking for missing values in percentage
(df.isnull().mean().sort_values(ascending=False) * 100).round(2)
```

```python
# Explore missingness vs. target variable
for col in df.columns:
    if df[col].isnull().any():
        df[col + '_missing'] = np.where(df[col].isnull(), 1, 0)
        print(col)
        print(df.groupby(col + '_missing')['Result'].value_counts(normalize=True))
        print("------->")
```

Based on the result above it shows that the missing values don't just occurs randomly rather it has a pattern. That means the missingness itself carries predictive power: people with missing symptom entries are mostly PENDING or NEGATIVE.

Dropping the missing values will lead to lost of important signals in the EDA section



```python
df.info()
```

```python
#Drop all columns that contain "_missing"
df_no_missing = df.loc[:, ~df.columns.str.contains("_missing")]
df_no_missing.info()
df = df_no_missing
```

# Univariate Analysis - Statistical Non Visual Analysis

```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("*"*10, col_name, "*"*10)
        print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

```python
numerical_univariate_analysis(numerical_df)
```

```python
numerical_df.columns
```

```python
# separating between discrete and continuous columns
discrete_num_cols = ['Sex', 'Chest_pain', 'Cough', 'Diarrhea',
                 'Fatigue_or_general_weakness', 'Fever',
                 'Headache', 'Thorax_(sore_throat)', 'Nausea',
                 'Runny_nose', 'Sore_throat_or_pharyngitis',
                 'Vomiting', 'Loss_of_Taste', 'Loss_of_Smell', 'Result']

continuous_cols = ['Birth_Year']
```

```python
discrete_num_df = df[discrete_num_cols]

print('Shape:', discrete_num_df.shape)
print('Columns:', discrete_num_df.columns)
```

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
```

```python
discrete_univariate_analysis(discrete_num_df)
```

# Bivariate Analysis - Visual Analysis

```python
df = df[df["Result"] != "INDETERMINATE"]
```

 ## *COVID-19 vs SEX*

```python
# Compare sex with COVID-19 test/result outcome
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Sex', hue='Result', palette='viridis')
plt.title("Sex vs COVID-19 Test Result")
plt.xticks(rotation=45)
plt.show()
```

```python
pd.crosstab(df['Result'], df['Sex'])
```

```python
ct = pd.crosstab(df['Result'], df['Sex'])
exc_no_pending = ct.drop(index='PENDING', errors='ignore')
pl = 'POSITIVE'
covid_rate_excluding_pending = (exc_no_pending.loc[pl] / exc_no_pending.sum(axis=0)) * 100
ct.loc['COVID Rate (Excl. Pending) (%)'] = covid_rate_excluding_pending
print(ct)
```
